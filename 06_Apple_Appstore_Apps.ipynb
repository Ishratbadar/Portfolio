```python
# Question 3. Which category has the highest average user rating?
# Calculate the average user rating for each genre
genre_avg_ratings = df.groupby('Primary_Genre')['Average_User_Rating'].mean()

# Sort the genres based on average user rating in descending order
sorted_genres_ratings = genre_avg_ratings.sort_values(ascending=False)

# Find the genre with the highest average user rating
highest_rating_genre = genre_avg_ratings.idxmax()

# Display the result
print("Genre with the highest average user rating:", highest_rating_genre)

# Plot the genres and their average ratings
plt.figure(figsize=(8, 6))
sorted_genres_ratings.plot(kind='bar')
plt.title('Average User Rating by Primary Genre')
plt.xlabel('Primary Genre')
plt.ylabel('Average User Rating')
plt.xticks(rotation=90)
plt.show()
```

**Figure-:** Shows the Distribution of all Categories according to the User Ratings 

```python
# Display the sorted genres and their average ratings
print("Primary Genres with Average User Rating (descending order):")
print(sorted_genres_ratings)
```

Answer 3: Weather Apps have the most Ratings, Games Apps have the 2nd highest Ratings and Photo & Video have the 3rd highest Ratings as shown in graph and in above table

```python
# Question 4. Which category has the highest average reviews?
# Calculate the average reviews for each genre
genre_avg_reviews = df.groupby('Primary_Genre')['Reviews'].mean()

# Sort the genres based on average user reviews in descending order
sorted_genres_reviews = genre_avg_reviews.sort_values(ascending=False)

# Find the genre with the highest average user reviews
highest_reviews_genre = genre_avg_reviews.idxmax()

# Display the result
print("Genre with the highest average reviews:", highest_reviews_genre)

# Plot the genres and their average reviews
plt.figure(figsize=(8, 6))
sorted_genres_reviews.plot(kind='bar')
plt.title('Average User Reviews by Primary Genre')
plt.xlabel('Primary Genre')
plt.ylabel('Average User Reviews')
plt.xticks(rotation=90)
plt.show()
```

**Figure-:** Shows the Distribution of all Categories according to the User Reviews

```python
# Display the sorted genres and their average reviews
print("Primary Genres with Average User Reviews (descending order):")
print(sorted_genres_reviews)
```

Answer 4: Weather Apps have the most User Reviews, Photo & Video Apps have the 2nd highest User Reviews and Finance have the 3rd highest User Reviews as shown in graph and in above table

```python
# Question 5. Which category has the highest average Price?
# Calculate the average Price for each genre
genre_avg_price = df.groupby('Primary_Genre')['Price'].mean()

# Sort the genres based on average user price in descending order
sorted_genres_price = genre_avg_price.sort_values(ascending=False)

# Find the genre with the highest average price
highest_price_genre = genre_avg_price.idxmax()

# Display the result
print("Genre with the highest average price:", highest_price_genre)

# Plot the genres and their average ratings
plt.figure(figsize=(8, 6))
sorted_genres_price.plot(kind='bar')
plt.title('Average price by Primary Genre')
plt.xlabel('Primary Genre')
plt.ylabel('Average price')
plt.xticks(rotation=90)
plt.show()
```
